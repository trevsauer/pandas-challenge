## Purchasing Analysis (Gender)

```python
# set up new dataframe that does not include repeats in SN
duplicate_SN = purchase_data_df.drop_duplicates(subset="SN", keep="first")
#shows total amount of SN, which does not include repeats 
Total_SN = duplicate_SN["Gender"].count()

Male_total = duplicate_SN["Gender"].value_counts()["Male"]

Female_total = duplicate_SN["Gender"].value_counts()["Female"]

No_SN = Total_SN - Male_total - Female_total 

No_SN_P = (No_SN / Total_SN) * 100

Male_P = (Male_total / Total_SN) * 100

Female_P = (Female_total / Total_SN) * 100 

#creates new data frame to display gender demographics
Gender_demographics = pd.DataFrame({"": ["Male", "Female", "Gender-Confused"], 
                                    "Percentages" : [Male_P, Female_P, No_SN_P], 
                                    "Total": [Male_total, Female_total, No_SN]})

#formatting data frame for cleanslinesssss
Gender_demographics["Percentages"] = Gender_demographics["Percentages"].map("{:.2f}%".format)
Gender_demographics = Gender_demographics.set_index('')
Gender_demographics
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#gender grouping dataframe
grouped_df = purchase_data_df.groupby(["Gender"])

#counts total purchases by SN
purchase_count = grouped_df["SN"].count()

#totals amount of purchases by gender
purchase_amount = grouped_df["Price"].sum()

#calculates average purchase price by gender 
avg_purchase_price = grouped_df["Price"].mean()

#gets rid of duplicates in data
duplicate = purchase_data_df.drop_duplicates(subset="SN", keep="first")

#groups data that isnt duplicated
duplicate_grouped = duplicate.groupby(["Gender"])

# sets variable to equal the prices in specialized grouped DF divieded by the respective SN count 
purchases_grouped = (grouped_df["Price"].sum() / duplicate_grouped["SN"].count())

#new data frame for gender based purchase analysis 
Gender_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                              "Average Purchase Price": avg_purchase_price,
                              "Total Purchase Value": purchase_amount,
                              "Avg Total Purchase per Person": purchases_grouped})

#formatting data frame to make prices have two decimals with $$$$ signs 
Gender_analysis["Average Purchase Price"] = Gender_analysis["Average Purchase Price"].map("${:.2f}".format)
Gender_analysis["Total Purchase Value"] = Gender_analysis["Total Purchase Value"].map("${:.2f}".format)
Gender_analysis["Avg Total Purchase per Person"] = Gender_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)

#shows data frame with correct headings & values
Gender_analysis = Gender_analysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
Gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#sets up bins based on age, lowest being under ten aka 0, and highest being 100, since nobody lives longer then that
bins = [0,10,15,20,25,30,35,40,100]
bin_headers = ["Under 10", "10 - 14", "15 - 19", "20 - 24", "25 - 29", "30 - 34", "35 - 39", "Over 40"]

#create new data frame and add the bins 
bin_df = purchase_data_df.copy()
bin_df["Age Bins"] = pd.cut(bin_df["Age"], bins, labels=bin_headers)
#bin_df.head()
#groups bins by age demographics
grouped_bins = bin_df.groupby(["Age Bins"])
#grouped_bins.head()
#counts the numbers in age demographic
bin_count = grouped_bins["SN"].count()
# gets total amount of age group stats for percentage calculation 
all_ages = purchase_data_df["SN"].count()
#calculates the percentage of players according to age demographics
age_percentage = (bin_count / all_ages) * 100 


#new DF to display the age demographic total counts and respective percentages 
Percentages_df = pd.DataFrame({"Total Count": bin_count, "Percentage of Players": age_percentage})

#formatting
Percentages_df["Percentage of Players"] = Percentages_df["Percentage of Players"].map("{:.2f}%".format)
Percentages_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#sets up bins based on age, lowest being under ten aka 0, and highest being 100, since nobody lives longer then that
#copied directly from last part, but created new DF for new bin organization 
bins = [0,10,15,20,25,30,35,40,100]
bin_headers = ["Under 10", "10 - 14", "15 - 19", "20 - 24", "25 - 29", "30 - 34", "35 - 39", "Over 40"]
new_bin_df = purchase_data_df.copy()
new_bin_df["Age Bins"] = pd.cut(new_bin_df["Age"], bins, labels=bin_headers)

#sets column to respective bin headers, and groups by age demographics  
column = pd.cut(new_bin_df["Age"], bins, labels=bin_headers)
binnerz = new_bin_df.groupby(["Age Bins"])
binnerz.head()
#run basic calculations for binnerzzz 
purch_count = binnerz["Age"].count()
av_purch_price = binnerz["Price"].mean()
total_purch = binnerz["Price"].sum()

#account for duplicates in binzzzz
binnerz_dup = purchase_data_df.drop_duplicates(subset="SN", keep="first")
binnerz_dup["Age Bins"] = pd.cut(binnerz_dup["Age"], bins, labels=bin_headers)
binnerz_dup = binnerz_dup.groupby(["Age Bins"])

dropped = (binnerz["Price"].sum() / binnerz_dup["SN"].count())

#formatting new DF 
Purchasing_analysis = pd.DataFrame({"Purchase Count": purch_count, "Average Purchase Price": av_purch_price, "Total Purchase Value": total_purch, "Avg Total Purchases Per Person": dropped})

Purchasing_analysis["Average Purchase Price"] = Purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
Purchasing_analysis["Total Purchase Value"] = Purchasing_analysis["Total Purchase Value"].map("${:.2f}".format)
Purchasing_analysis["Avg Total Purchases Per Person"] = Purchasing_analysis["Avg Total Purchases Per Person"].map("${:.2f}".format)
Purchasing_analysis = Purchasing_analysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchases Per Person"]]
Purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_group = purchase_data_df.groupby(["SN"])
SN_count = SN_group["Item ID"].count()
SN_total = SN_group["Price"].sum()
SN_avg = (SN_total / SN_count)

# new DF and formaTTT
SN_df = pd.DataFrame({"Purchase Count": SN_count, "Average Purchase Price": SN_avg, "Total Purchase Value": SN_total})

#formatting and sort in descending order
SN_df = SN_df.sort_values("Total Purchase Value", ascending = False)
SN_df["Average Purchase Price"] = SN_df["Average Purchase Price"].map("${:.2f}".format)
SN_df["Total Purchase Value"] = SN_df["Total Purchase Value"].map("${:.2f}".format)
SN_df = SN_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
SN_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Item_group = purchase_data_df.groupby(["Item ID", "Item Name"])
Item_count = Item_group["SN"].count()
Item_price = Item_group["Price"].sum()
Total_item_price = (Item_price / Item_count)
Item_purchase_value = (Total_item_price * Item_count)

#New DF with formatting swagger
Item_Summary = pd.DataFrame({"Purchase Count": Item_count, "Item Price":Item_price, "Total Purchase Value":Item_purchase_value})

#formatiiooone, aliging the purchase count column in descending order
Item_Summary = Item_Summary.sort_values("Purchase Count", ascending=False)
Item_Summary["Item Price"] = Item_Summary["Item Price"].map("${:.2f}".format)
Item_Summary["Total Purchase Value"] = Item_Summary["Total Purchase Value"].map("${:.2f}".format)
Item_Summary = Item_Summary[["Purchase Count", "Item Price", "Total Purchase Value"]]
Item_Summary
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# sort by descending order
Item_descend = purchase_data_df.groupby(["Item ID", "Item Name"])
descend_count = Item_descend["Gender"].count()
descend_sum = Item_descend["Price"].sum()
descend_avg = (descend_sum / descend_count)

Profitable = pd.DataFrame({"Purchase Count": descend_count, "Item Price": descend_avg, "Total Purchase Value": descend_sum})
#cleaner formatting
Profitable = Profitable.sort_values("Total Purchase Value", ascending=False)
Profitable["Item Price"] = Profitable["Item Price"].map("${:.2f}".format)
Profitable["Total Purchase Value"] = Profitable["Total Purchase Value"].map("${:.2f}".format)
Profitable.head()
```
